##### TF(t, d) = frequência do termo no documento / número total de tokens do documento

##### IDF(t) = log10(nº de documentos / nº de documentos que contêm o termo)

##### TF-IDF(t, d) = TF(t, d) × IDF(t)

##### cos(θ) = (A · B) / (||A|| × ||B||)

```python
from collections import Counter
from nltk import word_tokenize
from nltk.corpus import stopwords
import math
```

```python
path = "documents/document"
corpus = []

for i in range(1,4):
    with open(path + str(i) + ".txt", "r") as file:
        corpus.append(file.read())

def tokenizer(doc):
    return [word.lower() for word in word_tokenize(doc) if word.lower() not in stopwords.words("english")]

corpus_tokens = [tokenizer(doc) for doc in corpus]
print(corpus_tokens)

def getTF(words):
    total = len(words)
    count = Counter(words)

    for c in count:
        count[c] /= total

    return count

def idf_aux(word):
    total_docs = len(corpus_tokens)
    idf2 = 0
    for doc in corpus_tokens:
        if word in doc:
            idf2 += 1
    return math.log(total_docs/idf2, 10)

def getIdf(corpus_tokens):
    tokens = set([token for d in corpus_tokens for token in d])

    idf = {}
    for word in tokens:
        idf[word] = idf_aux(word)
    return idf

def getTFIDF(idf):
    tfidf_docs = []
    
    for i, doc in enumerate(corpus_tokens):
        print(f"Documento - {i}")
        tf = getTF(doc)
        tfidf = {}

        for word in tf:
            tfidf[word] = tf[word] * idf[word]
        tfidf_docs.append(tfidf)

    return tfidf_docs

def cosine_similarity(doc1, doc2):

    common = set(doc1.keys()) & set(doc2.keys())

    dot = sum(doc1[w] * doc2[w] for w in common)
    norm1 = math.sqrt(sum(v * v for v in doc1.values()))
    norm2 = math.sqrt(sum(v * v for v in doc2.values()))

    if norm1 == 0 or norm2 == 0:
        return 0
    
    return dot / (norm1 * norm2)

idf2 = getIdf(corpus_tokens)
tfidf = getTFIDF(idf2)

for i in range(len(tfidf)):
    for j in range(len(tfidf)):
        if i != j:
            sim = cosine_similarity(tfidf[i], tfidf[j])
            print(f"Doc {i} vs Doc {j}: {sim}")
```
