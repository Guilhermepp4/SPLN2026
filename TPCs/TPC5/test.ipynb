```python
import spacy

nlp = spacy.load("output/model-best")

texto = """O João foi ao Porto, para visitar a famosa Universidade do Porto, 
onde ele estudou Engenharia Informática. Ele também visitou a Torre dos Clérigos
e a Ribeira, que são pontos turísticos populares na cidade."""

doc = nlp(texto)

for ent in doc.ents:
    print(ent.text, ent.label_)
```
